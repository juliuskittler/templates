# Scatterplot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
plt.rcParams["figure.figsize"] = [15,10]

df = sns.load_dataset("tips")

p = sns.scatterplot(x="total_bill", y="tip", 
            hue="smoker", # Grouping variable that will produce elements with different colors.
            # style="smoker", # Grouping variable that will produce elements with different styles.
            # size="size", # Grouping variable that will produce elements with different sizes.
            data=df);

plt.title("Scatterplot: Y by X")
plt.show(p)
```

# Facet Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df = sns.load_dataset("tips")

p = sns.relplot(x="total_bill", y="tip", 
            col="time", # Categorical variables that will determine the faceting of the grid.
            hue="smoker", # Grouping variable that will produce elements with different colors.
            style="smoker", # Grouping variable that will produce elements with different styles.
            size="size", # Grouping variable that will produce elements with different sizes.
            height = 10,
            data=df);

plt.show(p)
```

# Line Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
plt.rcParams["figure.figsize"] = [15,10]

df = sns.load_dataset("flights")

p = sns.lineplot(x="year", y="passengers", 
            hue="month", # Grouping variable that will produce elements with different colors.
            data=df)

plt.title("Line Plot: Y by X")
plt.show(p)
```
