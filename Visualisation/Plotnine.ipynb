# Line Chart

```python
from plotnine import *
from plotnine.data import economics

(ggplot(economics) + 
    geom_line(aes(x = "date", y = "psavert"), color = "grey") + 
    geom_line(aes(x = "date", y = "uempmed"), color = "forestgreen") + 
    labs(title = "Line Chart: Y by X") +
    theme_bw())
```

# Bar Chart

```python
from plotnine import *
from plotnine.data import mpg

(ggplot(mpg, aes(x = "factor(cyl)")) + 
    geom_bar(alpha = 0.8, fill = "steelblue", color = "grey") +
    coord_flip() +
    labs(title = "Bar Chart: X") + 
    theme_bw())
```

# Scatterplot

```python
from plotnine import *
from plotnine.data import mpg

(ggplot(mpg) +
    aes(x = "cty", y = "hwy", color = "factor(cyl)") +
    geom_point() + 
    labs(title = "Scatterplot: Y by X", color = "cylinder") + 
    scale_color_brewer(type = "div", palette = 5) +
    theme_bw())
```

# Histogram

```python
from plotnine import *
from plotnine.data import mpg

(ggplot(mpg, aes(x = "displ")) + 
    geom_histogram(alpha = 0.8, fill = "steelblue", color = "grey", bins = 20) +
    labs(title = "Histogram: X") + 
    theme_bw())
```

# Density Plot

```python
from plotnine import *
from plotnine.data import mpg

(ggplot(mpg, aes(x = "displ")) + 
    geom_density(alpha = 0.8, fill = "steelblue", color = "grey") +
    labs(title = "Density Plot: X") + 
    theme_bw())
```
